```python
from qiskit import QuantumCircuit, Aer
from qiskit.visualization import plot_bloch_multivector
from qiskit.quantum_info import Statevector
import math
```

## Qubits Initialized to State $|011>$

```python
qc = QuantumCircuit(3)
qc.initialize(1,0)
qc.initialize(1,1)
qc.h(2)
# qc.h(1)
state = Statevector(qc)
plot_bloch_multivector(state)
```

```python
qc.draw(output='mpl')
```

# C-Rotation

```python
qc.cp(math.pi/2, 1, 2)
qc.draw(output='mpl')
```

```python
state = Statevector(qc)
plot_bloch_multivector(state)
```

```python
qc.cp(math.pi/4, 0, 2)
qc.draw(output='mpl')
```

```python
state = Statevector(qc)
plot_bloch_multivector(state)
```

# Full QFT

```python
qc.h(1)
qc.cp(math.pi/2, 0, 1)
qc.h(0)
qc.swap(0,2)
qc.draw(output='mpl')
```

```python
state = Statevector(qc)
plot_bloch_multivector(state)
```
