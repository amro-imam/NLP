```python
from qiskit import QuantumCircuit, Aer
from qiskit.visualization import plot_bloch_multivector
from qiskit.quantum_info import Statevector
```

## Qubit Initialized to State $|0>$

```python
qc = QuantumCircuit(1)
state = Statevector(qc)
plot_bloch_multivector(state)
```

```python
qc.h(0)
qc.draw(output='mpl')
```

```python
state = Statevector(qc)
plot_bloch_multivector(state)
```

## Qubit Initialized to State $|1>$

```python
qc = QuantumCircuit(1)
qc.initialize(1)
state = Statevector(qc)
plot_bloch_multivector(state)
```

```python
qc.h(0)
qc.draw(output='mpl')
```

```python
state = Statevector(qc)
plot_bloch_multivector(state)
```
